```python
# mask values in correlation heatmap
import numpy as np
mask = np.ones((4,4), dtype = int)

for i in range(len(mask[0])):
  for j in range(i):
    mask[i, j] = 0
# print(mask)
corr = iris.corr(numeric_only = True)
# annot=True â†’ displays the numeric values as text labels on the plot (commonly used in heatmaps).
sns.heatmap(corr, cmap = "coolwarm", annot=True, mask = mask)
```

```python
# identify outliers in petal length for sertosa using iqr method
setosa = iris[iris['species'] == 'setosa']
setosa.head()
```

```python
setosa.shape
```

```python
q1 = setosa['petal_length'].quantile(0.25)
q2 = setosa['petal_length'].median()
q3 = setosa['petal_length'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = []

for x in setosa['petal_length']:
    if x < lower_bound and x > upper_bound:
        outliers.append(x)

print(outliers)
```
